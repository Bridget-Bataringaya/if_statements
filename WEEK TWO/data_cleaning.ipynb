```python
#libraries for data manipulation
import os
import pandas as pd
import numpy as np

#you can now load the data set
df = pd.read_csv("data.csv")
```

```python
#display first few rows of the dataset
df.head(20)
```

```python
#data preprocessing or data cleaning
#bad data handling
#1.empty cells 
#2.wrong formats  30/05/2025     30052025(now this is a wrong format because it shaows no distinction between the data month and year)
#3.wrong data     age 23           -23,  123, 
#4.duplicates
```

```python
#empty values
#check for empty values
print(df.isnull().sum())
```

```python
#wrong formats
print(df.dtypes)
```

```python
#check for nn numeric values in numeric columns
for col in ["Duration","Pulse","Maxpulse","Calories"]:
    if not pd.to_numeric(df[col],errors="coerce").notnull().all():
        print(f"Non-numeric values found in column: {col}")
```

```python
#wrong data
#check for negative values in numeric column
for col in ["Duration","Pulse","Maxpulse","Calories"]:
    if (df[col] < 0).any():
        print(f"Negative values found in column: {col}")
```

```python
#Duplicates
duplicates = df.duplicated()
if duplicates.any():
    print("Duplicates found:")
    print(df[duplicates])
else:
    print("No duplicates found.")    
```

```python
#recommendations
#1. handle empty values(get the average of the empty cells and fill those cells)
#2.convert wrong formats
#3.correct wrong data
#4.delete the duplicated data(you delete the duplicated row not columns)
```

```python
#fill empty values with mean for numeric columns
for col in["Duration","Pulse","Maxpulse","Calories"]:
    df[col].fillna(df[col].mean(), inplace = True)
```
